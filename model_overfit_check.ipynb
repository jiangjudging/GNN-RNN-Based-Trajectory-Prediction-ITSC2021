```python
# 数据名称'stp0805am-0820am_v337_f1933.pyg'
dirpath = "/home/jiang/trajectory_pred/GNN-RNN-Based-Trajectory-Prediction-ITSC2021/stp_data_all"
# v337_pygs = find_files(dirpath,recursive=True,suffix=".pyg",prefix="stp0805am-0820am_v337")
# print(len(v337_pygs))
# print(v337_pygs)
# out_dir = create_new_dir(dirpath,"stp0805am-0820am_v337")

# for f in v337_pygs:
#     fname = f.split("/")[-1]
#     destpath = os.path.join(out_dir, fname)
#     shutil.move(f, destpath)

pygs = find_files(dirpath,recursive=True,suffix=".pyg")
scene_id_list = [i.split("/")[-1].split("f")[0][:-1] for i in pygs]
scene_id_set = list(set(scene_id_list))
print(len(scene_id_list)/len(scene_id_set))

```

```python
from sklearn.model_selection import train_test_split
random.seed(0)

f1_train,f1_test = train_test_split(scene_id_set,test_size=0.1,shuffle=True)
f1_val,f1_test = train_test_split(f1_test,test_size=0.5,shuffle=True)

print(len(f1_train))
print(len(f1_val))
print(len(f1_test))
print(f1_test)
```

```python
from tqdm import tqdm
train_dir = "/home/jiang/trajectory_pred/GNN-RNN-Based-Trajectory-Prediction-ITSC2021/ngsim_data_id/ngsim_data_id_train"
val_dir = "/home/jiang/trajectory_pred/GNN-RNN-Based-Trajectory-Prediction-ITSC2021/ngsim_data_id/ngsim_data_id_val"
test_dir = "/home/jiang/trajectory_pred/GNN-RNN-Based-Trajectory-Prediction-ITSC2021/ngsim_data_id/ngsim_data_id_test"
dirs = [train_dir,val_dir,test_dir]
dts = [f1_train,f1_val,f1_test]
for dt,out_dir in zip(dts,dirs):
    for prefix in tqdm(dt):
        id_files = find_files(dirpath,recursive=True,suffix=".pyg",prefix=prefix)
        for id_file in tqdm(id_files):
            destpath = os.path.join(out_dir,id_file.split('/')[-1])
            shutil.copy(id_file,destpath)
```
